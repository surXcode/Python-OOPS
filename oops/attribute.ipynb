# How to fetch/accessing the property:--

```python
# Accessing properties by using class


class Demo:
    a = 10
    b = 20

ob1 = Demo()
ob2 = Demo()

print(Demo.a)
print(Demo.b)
print('=' * 15)
#Accessing properties by using object
print(ob1.a)
print(ob1.b)
print('=' * 15)
print(ob2.a)
print(ob2.b)
print('=' * 15)
# Modifying the properties:
Demo.a = True
print("After modify")
print(Demo.a)
print(ob1.a)
print(ob2.a)
print('=' * 15)
print("Modify by object")

ob1.a = 40
print(ob1.a)-0
print(ob2.a)
print(Demo.a)
```

# Types of properties/states/attributes/data/members:-

# There are two types of properties:

1. class property/member:- 

These are the members of class which  are same for all object.

Example: School, name , address, terms & condition.

2. object property/member:

These are the properties which are different for all the objects.

Example: Student---> name, roll number, addr, aadhar, phno.



```python
class School:
    name = "KNICE"
    Address = "Sultanpur"
    Principle_name = "Suraj Singh"
s1 = School()
s2 = School()
s3 = School()
# creating object properties
s1.sname = "Anuj"
s1.addr = "Pratapgarh"
s1.phno = '87654321'
print(s1.sname)
print(s1.addr)
print(s1.phno)
print('=' * 35)
s2.sname = "Ashvik"
s2.addr = "Mumbai"
s2.phno = '9876543212'
print(s2.sname)
print(s2.addr)
print(s2.phno)
print('=' * 35)
s3.sname = "Suraj"
s3.addr = "Sultanpur"
s3.phno = '786542157'
print(s3.sname)
print(s3.addr)
print(s3.phno)
```
