# 2. Method Overriding:

- changing the implementation of parent class method inside child class method according to the requirement of child class is known as method overriding.

-   To perform  method overriding -

   - Inheritance should be done 
   - Method signature should be same
   - Method type should be same

# Method signature:
 
- Method signature = method definition + parameter 

```python
class Animal():

    def sound(self):
        print("animal make sound")


class Human(Animal):

    def sound(self):
        print("Human speaks....")


class Dog(Animal):
    def sound(self):
        print("Dog barks...")


ob1 = Dog()
ob2 = Human()
ob1.sound()
ob2.sound()
```

# Operator overloading:

- It allows the operator to work on the user defined object/(object of the user different class).

- By using some magic methods /dunder method.

- operator works on inbuilt class.

- 

```python
class Arithmetic:

    def __init__(self, val):
        self.a = val

    def __add__(self, other):
        return self.a + other.a

    def __sub__(self, other):
        return self.a - other.a


ob1 = Arithmetic(10)
ob2 = Arithmetic(20)
print(ob1 + ob2)
print(ob1 - ob2)
```

# âš™ï¸ What is Operator Overloading in Python?

- Giving special meaning to operators (+, -, *, >, ==, etc.) for user-defined objects.

- Python lets you redefine how operators behave for your class using special (magic/dunder) methods.

# __str__() vs __repr__()

- Both are special (dunder) methods used to define how an object is shown as a string.

   -   __str__()
   -   __repr__()

ğŸ”¹ __str__()

- Human-readable, nice, clean output (for users)

- Called by: print(obj) and str(obj)

- Meant to be friendly display

ğŸ”¹ __repr__()

- Developer-readable, detailed, unambiguous output (for debugging)

- Called by: typing object in shell, repr(obj)

- Should ideally be something that can recreate the object

#  Interview Line

- __str__ is for readable display, while __repr__ is for precise representation of the object for developers.

```python
# ğŸ”¹ __str__()


class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"name is {self.name} , age is {self.age}"

    def __repr__(self):
        return f"Student({self.name} , age is {self.age})"


s1 = Student("Suraj Singh", 22)
print(s1)
```

```python
# Indexing and Updation:


class Collection:

    def __init__(self, val):
        self.list1 = val

    def __getitem__(self, index):
        return self.list1[index]

    def __setitem__(self, index, new):
        self.list1[index] = new

    def __len__(self):
        return len(self.list1)



ob1 = Collection([10, 20, 30])
print(ob1[0])
ob1[0] = 100
print(ob1.list1)
print("Length is :" ,len(ob1))
```
