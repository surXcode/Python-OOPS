```python
class Math:
    def add(self, a, b=0, c=0):
        return a + b + c

m = Math()

print(m.add(5, 10))      
print(m.add(5, 10, 20))  
print(m.add(5))          
```

```python
class Teacher:
    def role(self):
        print("Teaches students")

class Principal(Teacher):
    def role(self):  
        print("Manages the school and student")

p = Principal()
p.role()
```
