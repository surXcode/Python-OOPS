# Encapsulation

- It is process of wrapping a data in a single 

- This pillar is use to provide security to our data

- it restricts direct access to the data.

- We provide security by the help of access specifier

# There are thre types of access specifier:

  - Public

  - Protected

  - Private

# Public:

- Public access specifier allows the method and data to be accessed outside of the class also.

# Protected access specifier:

- We use single under score before the data or method to make it as proctected access specifier.

- These can be accessed inside the class and outside of the class also.

 * Syntax:
   
   - class ClassName:
       - 

# Private access specifier:

- We use double under score before the data and method to make it private access specifier. 

- Private can not be access inside the child class and outside of the class.

* Real-life idea:

- ATM machine — you can use it, but can’t access the internal system.

# Property decoratror:(without parenthesis):

- It allows us to use the method as property.

- without parenthesis() call the method.

```python
class Person:
    age = 20

    @property
    def height(self):
        print("Height is 152 cm")


p1 = Person()
p1.height
```

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    @property
    def show(self):
        return f"Name is {self.__name}, Age is {self.__age}"

    @property
    def name(self):
        return self.__name

    @property
    def age(self):
        return self.__age

    @name.setter
    def name(self, new):
        self.__name = new

    @age.setter
    def age(self, new):
        self.__age = new

    @age.deleter
    def age(self):
        del self.__age


# Object creation
p1 = Person("Tarun", 23)

# Access age
print(p1.age)
```
