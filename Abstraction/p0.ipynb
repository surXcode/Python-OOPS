# Abstraction:

- Hiding the internal implementation by providing the neccessary functionalities to the user is known as abstraction.

# Three things are required:

1. Abstract class

2. Abstract method

3. Concrete class

# 1. Abstract class:

- Abstract class is a child class of ABC class(abstract base class)

- Abstract class should contain atleast one abstract method

- We can not create object from abstract class can not be intantiated.

- ABC class is based abstract presend in ABC module.

# ✅ Syntax

- from abc import ABC, abstractmethod

- class Shape(ABC):
 -   pass

# 2. Abstract method:

- It is method which is declared/defined but not implemented.

- Abstract method decorator is used to before defining the abstract method. 

- this decorator presend in ABC module.

# ✅ Syntax

- from abc import ABC, abstractmethod

- class Shape(ABC):

 -   @abstractmethod
 -   def area(self):
  -      pass

# 3. Concrete class

- concrete class is child class

* We can create the object for concrete class:

# Syntax

- Concrete class (normal)
- class ClassName:
 -   def method(self):
   -     pass

# Syntax

- from abc import ABC, abstractmethod

- class Shape(ABC):

   - @abstractmethod
   - def area(self):
    -    pass


```python
# abstract
from abc import ABC,abstractmethod

class Vechicle(ABC):

    @abstractmethod
    def start_engine(self):
        pass
    # @abstractmethod
    # def check_light(self):
    #     pass
    @abstractmethod
    def stop_engine(self):
        pass
class Car(Vechicle):
    def start_engine(self):
        print("Engine has start")
    def stop_engine(self):
        print("Engine off")
c1 = Car()
c1.start_engine()
```

# Types of relation between classes:

# Two types of relation

# 1. Is -A relation(Inheritance)
# 2. Has -A relation

# 1. Is -A relation(Inheritance)

- One class is a specialized form of another class.


# 2. Has -A relation:

- One class contains a reference to another class.

- It is a type of relation where object of one class will be the property of anothder class.

# There are two types of Has-A relation:

1. Aggregation 

- object are  independent in class

- loosely coupled

- class is destroy it will not afect the object

2. Comosition

- Object are dependent on class

- tightly coupled

# Difference between:


 | Feature               | Aggregation        | Composition             |
|-----------------------|--------------------|-------------------------|
| Relationship strength | Weak               | Strong                  |
| Object lifetime       | Independent        | Dependent               |
| Object creation       | Outside the class  | Inside the class       |
| Ownership             | Shared             | Owned                   |
| Example idea          | Car uses an engine | Car contains its engine |


```python
class Address:                # Aggregation Example

    def __init__(self, state, loc, pin):
        self.state = state
        self.loc = loc
        self.pin = pin


addr = Address("Up", "Sultanpur", 228001)


class Person:

    def __init__(self, name, age, addr):
        self.name = name
        self.age = age
        self.address = addr


p1 = Person("Suraj", 22, addr)      # object of address
print(p1.name)
print(p1.age)
print(p1.address.state)
print(p1.address.loc)
print(p1.address.pin)
```
