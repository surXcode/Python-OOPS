1. Single level inheritance: 
 
- Here, properties are transferred from a single parent class to a single child class, but the reverse is not possible.

# Syntax:
- class Parent:

                    ----------
                    |   SB   |
                    ----------
- class Child(Parent):
                     
                    ----------
                    |   SB   |
                    ----------







```python
class Main:
    a = 100
    b = "hello"


class Demo:
    a = 200
    b = "Bye"


class New(Demo):
    c = "This is child class"


ob1 = Main()
ob2 = Demo()
ob3 = New()

print(ob1.a)  
print(ob2.b)   
print(ob3.c)   
```

2. Multilevel Inheritance:

- In multilevel inheritance, one class inherits from another class, and then another class inherits from that class.

# Syntax: 
- class A:          # Parent Class
    pass

- class B(A):       # Child of A
    pass

- class C(B):       # Child of B (Grandchild of A)
    pass


# Diagram:
- A   ‚Üí   Parent Class
 ‚Üë
 |
- B   ‚Üí   Child of A
 ‚Üë
 |
- C   ‚Üí   Child of B (Grandchild of A)

```python
class First:
    a = 10
    b = 20


class Second(First):
    x = "hii"
    y = 'Hello'


class Third(Second):
    j = True
    k = False


ob1 = Third()
print(ob1.a)
print(ob1.b)
print(ob1.x)
print(ob1.y)
print(ob1.j)
print(ob1.k)
```

 # Method resolution order: 

- It is the order in which python look for a attribute or method when it calls it object .

- It searches inside the child class FIRST.

- It searches the parent class in a consistent mannner from left to right 

- MRO = the order in which Python searches classes for a method or attribute.

- When you call a method on an object, Python doesn‚Äôt randomly pick where to look. It follows a strict order called  the Method Resolution Order.

- This becomes very important in Multiple Inheritance.

#  Importance of MRO (Method Resolution Order):

MRO is important because it decides which method Python executes in multiple inheritance.

üîπ 1) Removes Confusion in Multiple Inheritance

If two parent classes have the same method, MRO clearly decides which one runs.

üîπ 2) Solves the Diamond Problem

Prevents the top parent from being called multiple times and keeps a consistent order.

üîπ 3) Makes super() Work Properly

super() follows the MRO path, not the immediate parent ‚Äî enabling clean method chaining.

üîπ 4) Predictable Method Lookup

Python searches classes in a fixed, known order ‚Üí no randomness, fewer bugs.

üîπ 5) Essential for Large OOP Projects

With many base classes, MRO ensures safe, structured inheritance behavior.

```python
class A:           # MRO example
    a = 10
    b = "hii"


class B:
    a = 100
    b = "bye"


class C:
    a = True
    b = False


class child(C, B, A):
    pass


c1 = child()
print(c1.a)
print(c1.b)
```

```python
class Addition:

    @staticmethod
    def add(a, b):
        return a + b


class Subtraction:

    @staticmethod
    def sub(a, b):
        return a - b


class Product:

    @staticmethod
    def pro(a, b):
        return a * b


class Calculator(Addition, Subtraction, Product):

    @staticmethod
    def div(a, b):
        return a / b


ob1 = Calculator()
print(ob1.add(100, 20))
print(ob1.sub(100, 20))
print(ob1.pro(100, 20))
print(ob1.div(100, 20))
```

# Hybrid Inheritance:- 

- It is the combination of two or more than two types of inheritance.

- 

# Syntax:

       Class A
      /       \
 Class B     Class C
      \       /
       \     /
        Class D


```python
class Math:
    a = "this is math class "


class Bio(Math):
    a = "this is bio class "


class Calculation(Bio):
    pass


class chemistry(Calculation):
    pass


class physics(Calculation):
    pass

print(Math.a)
print(Bio.a)
print(Calculation.a)
print(chemistry.a)
print(physics.a)
```

# Constructor chainnig:-

- It is a process of calling parent class constructor inside the child class constructor.

- It is phenomenon of calling the parent class constructor inside the child class.

- if we want to add extra properties for the child class onbject then we should use constructor chaining.

- We can do so by using super function or class name.

- constructor call should be done inside the constructor of the child class.



# Syntax:

                +-------------------------+
                |        CLASS            |
                |   -------------------   |
                |   |   Constructor   |   |
                |   |   __init__()    |   |
                |   -------------------   |
                |          |              |
                +----------|--------------+
                           |
                           | Object Creation
                           v
                +-------------------------+
                |        OBJECT           |
                |  (Memory Allocated)     |
                |   Properties Initialized |
                +-------------------------+





```python
class Demo:

    def __init__(self):
        print("This is demo class constructor")


class Primary(Demo):

    def __init__(self):
        super().__init__()
        print("This is primary class constructor")


ob1 = Primary()


# Constructor calling by using class
class A:

    def __init__(self):
        print("This is a constructor of A")


class B:

    def __init__(self):
        print("This is a constructor of B")


class C:

    def __init__(self):
        print("This is a constructor of C")


class Main(A, B, C):

    def __init__(self):
        C.__init__(self)
ob1 =Main()
```

# There are two way perform constructor chaining:

- Super()

- By Using class name.

- Super():

super() is a built-in function that lets you call methods of the parent class without writing the parent‚Äôs name.

# Syntax:
  
  - super().method_name()

- By Using class name:

You directly call the parent class method using its class name.

# Syntax:

   - Parent.method_name(self)


```python
class GraminBank:         # Constructor chainnig by using super()
    def __init__(self, name, addr):
        self.name = name
        self.addr = addr


class SBI(GraminBank):                    # inherit from Bank

    def __init__(self, name, addr, email, phno):
        super().__init__(name, addr)
        self.email = email
        self.phno = phno


# Pass all required arguments
c1 = SBI('Suraj', "Noida", "suraj@gmail.com", 9876543210)

print(c1.name)
print(c1.addr)
print(c1.email)
print(c1.phno)
```

# Method chaining:

- It is a process of calling multiple method by using a single object in single line of expression.

- In order to achieve method chaing the methos present inside the class should return object reference/self.

# For chaining to work, a method must return:

- self (same object), or

- another object that has the next method.

# Class Method Chaining Template:

- class ClassName:
-    def method1(self, args):
 -       # do something
 -       return self

-    def method2(self, args):
 -       # do something
  -      return self

# Syntax:

   - object.method1().method2().method3()-----methodn()




```python
class Person:

    def __init__(self):
        self.name = None
        self.age = None

    def set_name(self, name):
        self.name = name
        return self

    def set_age(self, age):
        self.age = age
        return self

    def show(self):
        print(f"Person name is {self.name}")
        print(f"Person age is {self.age}")
        return self


p1 = Person()
p1.set_name("Suraj").set_age(23).show()

```
